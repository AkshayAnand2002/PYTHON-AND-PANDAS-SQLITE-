# LOADING DATA FROM SQLITE DATABASES INTO DATAFRAME

```python
import pandas as pd
import sqlite3 as sq3
```

```python
con=sq3.connect("movies.db")##connect to the required database.
```

```python
con.execute("SELECT * FROM sqlite_master").fetchall()
```

```python
##FROM ABOVE COMMAND WE GET ALL ELEMENTS OF SQLITE DATABASE MOVIES.DB
```

```python
##SO FROM ABOVE COMMAND WE HAVE LOADED FULL MOVIES TABLE INTO PANDAS DATAFRAME.
```

```python
##NOW WE WRITE A SIMPLE SQL QUERY TO GET ALL ELEMENTS OF "Movies" TABLE IN MOVIES.DB DATABASE.
```

```python
pd.read_sql("SELECT * FROM Movies",con)###TO GET ALL ELEMENTS OF Movies TABLE IN THE DATABASE.
```

```python
##TO GET DIFFEREENT DATATYPES OF DIFFERENT ELEMENTS OF MOVIES TABLE.
```

```python
pd.read_sql("SELECT * FROM Movies",con).info()
```

```python
##SOME PROBLEM IN DATATYPE AS RELEASE_DATE SHOULD OF DATATIME DATATYPE.
```

```python
df=pd.read_sql("SELECT * FROM Movies",con,index_col="id",parse_dates="release_date")
df
##THROUGH ABOVE WE SET INDEX OF TABLE AS ID INSTEAD OF INDEXIN INTEGERS 0,1,2 ETC AND release_date has
##changed because of parse_dates parameter.
##ALSO FROM ABOVE COMMAND WE HAVE CORRECTED THE RELEASE_DATE DATATYPE.
```

```python
df.info()
```

```python
###SEE DATATYPE IS CORRECTED.
```

```python
genres=pd.read_sql("SELECT * FROM GENRES",con,index_col="id")
genres
##PRINTING ALL ELEMENTS FROM GENRES TAKING INDEX AS ID.
```
