```python
import pandas as pd
import sqlite3 as sq3
```

```python
con=sq3.connect("movies.db")
```

```python
##TO GET ALL TABLES PRESENT IN THE MOVIES.DB DATABASE
con.execute("SELECT * FROM sqlite_master").fetchall()
```

```python
##TO GET ALL COLUMNS FROM MOVIES TABLE IN THE PANDAS DATAFRAME.
pd.read_sql("SELECT * FROM Movies",con)
```

```python
###,con passed in above and previous commands for connection object.
```

```python
##TO WRITE COMMANDS OR SQL QUERIES IN MORE THAN ONE LINE WE USE BACKSLASH.
pd.read_sql("SELECT * \
            FROM Movies",con)
```

```python
##SAME RESULT OF ABOVE AS PREVIOUS SQL QUERY IN 1 LINE.
```

```python
pd.read_sql("SELECT id,revenue,release_date FROM Movies",con)
```

```python
##we have an indexing of 0,1,2 so on we can remove then and make any column as index.
pd.read_sql("SELECT id,revenue,release_date FROM Movies",con,index_col="id")
```

```python
pd.read_sql("SELECT SUM(revenue) FROM Movies",con)##SUM OF REVENUES FROM MOVIES TABLE IN MOVIES DATABASE.
```

```python
con.execute("SELECT SUM(revenue) FROM Movies").fetchall()
##USING .EXECUTE METHOD TO GET TOTAL REVENUE .
```

```python
##THE FIRST ELEMENT OF .EXECUTE METHOD WILL BE THE TOTAL REVENUE.
con.execute("SELECT SUM(revenue) FROM Movies").fetchall()[0][0]
```

```python
#COUNT NO. OF NON-MISSING VALUES OF TITLE FROM MOVIES.
pd.read_sql("SELECT COUNT(TITLE) FROM Movies",con)
```

```python
###COUNT NO. OF NON-MISSING VALUES FROM belong_to_collection_name FROM MOVIES TABLE.
pd.read_sql("SELECT COUNT(belongs_to_collection_name) FROM Movies",con)
```

```python
##THE ABOVE IS 15 AND NOT 18 AS 3 VALUES ARE NONE OR NULL.
```

```python
##COUNT ALL ROWS FROM MOVIES TABLE.
pd.read_sql("SELECT COUNT(*) FROM Movies",con)
```

```python
##GET THE AVERAGE FROM BUDGET COLUMN.
pd.read_sql("SELECT AVG(budget) FROM Movies",con)
```
