```python
import pandas as pd
import sqlite3 as sq3
```

```python
con=sq3.connect("movies.db")
```

```python
###TO GET TO KNOW NAMES AND PROPERTIES OF ALL TABLES IN SQL DATABASE CONNECTED.
con.execute("SELECT * FROM sqlite_master").fetchall()
```

```python
df=pd.read_sql("SELECT Prod.id,Prod.comp_name,Movies.revenue,Movies.title \
            FROM Prod \
            LEFT JOIN Movies \
            ON Prod.id = Movies.id",con)
df
```

```python
###IN THE ABOVE LEFT JOIN WE WILL GET THE SAME ANSWER AS INNER JOIN AS NO KEY IS REPEATING HERE.
###BY DEFAULT IF WE WRITE 'JOIN' THEN IT MEANS 'INNER JOIN' ONLY.
```

```python
df1=pd.read_sql("SELECT Prod.id,Prod.comp_name,Movies.revenue,Movies.title \
            FROM Prod \
            INNER JOIN Movies \
            ON Prod.id = Movies.id",con)
df1
```

```python
df2=pd.read_sql("SELECT Prod.id,Prod.comp_name,Movies.revenue,Movies.title \
            FROM Prod \
            JOIN Movies \
            ON Prod.id = Movies.id",con)
df2
```

```python
df.groupby("comp_name").revenue.sum().sort_values(ascending=False)
###SORT IN DESCENDING ORDER BY CALCULATING SUM OF REVENUES FOR PARTICULAR COMPANY NAMES
###THAT IS IF A COMPANY HAS PRODUCED MORE THAN 1 MOVIE THEN CALCULATE SUM OF REVENUES IN ALL MOVIES PRODUCED BY 
### COMPANY AND DO SO FOR ALL COMPANIES AND THEN SORT THE SUM BASED ON DESCENDING ORDER.
```

```python
##INTERESTED IN FINDING THE COMPANY NAMES INVOLVED IN PRODUCING TITANIC MOVIE.
pd.read_sql("SELECT Prod.comp_name \
            FROM Prod \
            LEFT JOIN Movies \
            On Prod.id = Movies.id \
            WHERE Movies.title = 'Titanic'",con)
```

```python
df4=pd.read_sql("SELECT Genres.id,Genres.genre_name,Movies.revenue,Movies.title \
                FROM Genres \
                LEFT JOIN Movies \
                ON Genres.id = Movies.id",con)
df4
```

```python
df4.groupby("genre_name").revenue.sum().sort_values(ascending=False)
###THE ABOVE COMMAND GROUPS BY REVENUE AND SORTS BY SUM OF REVENUE IN DESCENDING ORDER.
```
